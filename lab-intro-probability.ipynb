**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
"""
We are computing a discreat value, 
The goal is to know compute the cdf the % for the passerger to have a seat 
With 460 seats sells / 450 max seats this distrib model is binomial
"""

from scipy.stats import binom

# 1. Définir les paramètres identifiés
n = 460  # tickets vendus
p_present = 0.97  # probabilité qu'un passager se présente (1 - 0.03)

# 2. Appliquer la logique : P(X <= 450)
# On utilise cdf car c'est une probabilité cumulée (de 0 à 450)
proba_to_have_seat = binom.cdf(450, n, p_present)

print(f"the probability to be seat : {proba_to_have_seat:.4f}")
```

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
"""
We are computing the nbre of call after 2 => discreat value ; geometric
to know proba to resolve in a time the complain cdf
With 0.3 at the first attempt
-> cdf ;  Géométrique ; started at 2
"""
from scipy.stats import geom
p = 0.3
prob = geom.sf(2, p)

print(f" the probability that the representative needs to make at least 3 attempts is : {prob}")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
"""
We are computing the number of visit at least => discreat value 
to saturate the server (550max for 500 on average) => poisson
-> cdf poisson.cdf(limite, mu) ;  poisson
"""
from scipy.stats import poisson

mu = 500
limite = 550
prob_overwhelmed = 1 - poisson.cdf(limite, mu) 
# Probabilité d'être submergé : P(X > 550)
# Ou utiliser la survival function (sf) qui est 1 - cdf
# prob_overwhelmed = poisson.sf(limite, mu)
print(f"the probability that the server being overwelmed is during 1h: {prob_overwhelmed:.3f}")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
#For a day
mu = 500 * 24
limite = 550 * 24
prob_overwhelmed = 1 - poisson.cdf(limite, mu)
print(f"the probability that the server being overwelmed is during 24h: {prob_overwhelmed:.10f}")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
"""
We are computing the time between 2 arrivals => continus value 
to be inferior or equal at 5 min => exponetielle
-> cdf ;  exponentioelle
"""
from scipy.stats import expon

# Moyenne d'attente (1/lambda)
average = 10 

# Temps limite souhaité
time_limit = 5

# Calcul de P(X <= 5)
# On utilise la CDF car on veut "moins de 5 minutes"
prob = expon.cdf(time_limit, scale=average)

print(f"Probability that the next customer will arrive within the next 5 min : {prob * 100:.2f}%")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
"""
We are computing the time break => continus value 
for 15 min without arrival => exponetielle
-> survival fonction ;  exponentielle
"""

# Moyenne d'attente (1/lambda)
average = 10 

# Temps limite souhaité
break_allowed = 15

# Calcul de P(X <= 5)
# On utilise la CDF car on veut "moins de 5 minutes"
prob = expon.sf(break_allowed, scale=average)

print(f"Probability to take a break is: {prob * 100:.2f}%")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
"""
We are computing the weight for a bird between bird => continus value 
for random mean weight bird of 150g => normal
-> P(140 < X < 160) = CDF(160) - CDF(140) = 10 ;  normal
"""
from scipy.stats import norm

mean = 150
std = 10

# On calcule la probabilité cumulative jusqu'à 160 et on retire celle jusqu'à 140
probability = norm.cdf(160, mean, std) - norm.cdf(140, mean, std)

print(f"La probabilité est de : {probability:.3f}")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
"""
We are computing the electronics componenent reaching 30h => continus value 
for mean lifetime of 50h => expo
->  ;  epo
"""
from scipy.stats import expon

# Paramètres
mean_lifetime = 50 

# "Within the first 30 hours" -> P(X <= 30)
# En Python, scale correspond à la moyenne pour la loi exponentielle
probabilite = expon.cdf(30, scale=mean_lifetime)

print(f"La probabilité de panne avant 30h est : {probabilite:.3f}")
```
